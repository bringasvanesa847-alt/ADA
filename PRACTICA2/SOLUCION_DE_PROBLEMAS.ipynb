### Eliminacion de duplicados en CSV
Este programa elimina registros duplicados basandose en el campo transaction_id, conservando unicamente la primera aparicion.

### Descripcion
Se recorre el archivo fila por fila y se utiliza un conjunto (set) para almacenar los identificadores ya vistos. Si un identificador no ha sido visto, se agrega al resultado.


```python
import csv
import os

def remove_duplicates_csv(input_file, output_file):
    """
    Elimina registros duplicados de un archivo CSV basandose en el campo
    'transaction_id', conservando unicamente la primera aparicion.

    Complejidad temporal:
    - O(n), donde n es el numero de filas del archivo.
      Esto se debe a que se recorre el archivo una sola vez y las operaciones
      dentro del ciclo son de tiempo constante en promedio.

    Complejidad espacial:
    - O(n) en el peor caso, debido al almacenamiento de los identificadores
      unicos en un conjunto (set).
    """

    seen = set()  
    # Tiempo: O(1)
    # Memoria inicial: O(1)
    # Este conjunto almacenara los transaction_id ya vistos

    try:
        # Verificacion de existencia del archivo
        if not os.path.exists(input_file):
            print("Error: el archivo de entrada no existe")
            return
        # Tiempo: O(1)

        with open(input_file, "r", newline="") as infile, \
             open(output_file, "w", newline="") as outfile:
            # Apertura de archivos
            # Tiempo: O(1)
            # Memoria: O(1)

            reader = csv.DictReader(infile)
            # Tiempo: O(1)
            # Prepara la lectura sin cargar todo el archivo en memoria

            if not reader.fieldnames:
                print("Error: el archivo no tiene encabezados")
                return
            # Tiempo: O(1)

            writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
            # Tiempo: O(1)

            writer.writeheader()
            # Tiempo: O(1)

            for row in reader:
                # Este ciclo se ejecuta n veces
                # Tiempo total del ciclo: O(n)

                tid = row.get("transaction_id")
                # Tiempo: O(1)
                # Acceso a diccionario

                if tid and tid not in seen:
                    # tid: O(1)
                    # busqueda en set: O(1) en promedio

                    seen.add(tid)
                    # Insercion en set: O(1) en promedio
                    # Memoria acumulada: O(n)

                    writer.writerow(row)
                    # Escritura de una fila: O(1)

        print("Archivo creado correctamente")
        print("Ruta:", os.path.abspath(output_file))
        # Tiempo: O(1)

    except Exception as e:
        print("Error:", e)
        # Manejo de errores: O(1)


if __name__ == "__main__":
    input_file = "transaction_data.csv"
    output_file = "output.csv"

    remove_duplicates_csv(input_file, output_file)
```

### Complejidad
- Tiempo: O(n), ya que se recorre el archivo una sola vez
- Memoria: O(n), debido al almacenamiento de los identificadores unicos

### Resultado
El archivo generado contiene unicamente transacciones unicas, manteniendo el orden original.
