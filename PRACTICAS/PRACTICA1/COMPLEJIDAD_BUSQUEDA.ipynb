# ALGORITMOS DE BUSQUEDA

Prompt: Cómo puedo compilar y ejecutar código C++ desde un notebook?

```python
%%bash
# Indica que las siguientes líneas se ejecutarán en la terminal (bash).
# Se usa normalmente en Jupyter Notebook o Google Colab.

g++ COMPLEJIDAD_BUSQUEDA.cpp -O2 -o busqueda
# g++ es el compilador de C++.
# COMPLEJIDAD_BUSQUEDA.cpp es el archivo fuente que se va a compilar.
# -O2 activa un nivel de optimización del compilador para que el programa sea más eficiente.
# -o busqueda indica que el archivo ejecutable generado se llamará "busqueda".

./busqueda > resultados.csv
# ./busqueda ejecuta el programa compilado.
# > resultados.csv redirige la salida del programa a un archivo llamado "resultados.csv".
# En lugar de mostrarse en pantalla, los resultados se guardan en ese archivo.
```

```python
import pandas as pd

df = pd.read_csv("resultados.csv")
df.columns = df.columns.str.strip()

print(df.head())
```

# GRÁFICA DE TIEMPO

Esta gráfica muestra el tiempo promedio de ejecución (en nanosegundos) de los tres algoritmos de búsqueda según el tamaño del arreglo.

En el eje horizontal está el tamaño del arreglo (desde 1,000 hasta 100,000 elementos aproximadamente).
En el eje vertical está el tiempo promedio en nanosegundos (ns).

### ¿Qué se observa?
Búsqueda Lineal:
El tiempo aumenta conforme crece el tamaño del arreglo.
Esto sucede porque revisa elemento por elemento hasta encontrar el valor.
Mientras más grande sea el arreglo, más comparaciones necesita hacer.

Búsqueda Binaria:
Su tiempo se mantiene casi constante incluso cuando el arreglo crece mucho.
Esto ocurre porque en cada paso divide el arreglo a la mitad.
Por eso su crecimiento es muy pequeño comparado con los otros métodos.

Jump Search:
Su tiempo aumenta más que la binaria y en tu gráfica es el que más crece.
Esto sucede porque avanza por bloques (saltos) y luego hace una búsqueda lineal dentro del bloque.
A medida que el arreglo crece, también aumenta el tamaño del salto y el número de comparaciones.

```python
import matplotlib.pyplot as plt

plt.plot(df["Tamano"], df["Linear"])
plt.plot(df["Tamano"], df["Binary"])
plt.plot(df["Tamano"], df["Jump"])

plt.xlabel("Tamaño del arreglo")
plt.ylabel("Tiempo (ns)")
plt.legend(["Linear", "Binary", "Jump"])
plt.show()
```

# GRÁFICA DE TIEMPO

Esta gráfica muestra el consumo de memoria (en bytes) conforme aumenta el tamaño del arreglo.

En el eje horizontal está el tamaño del arreglo (desde 1,000 hasta 100,000 elementos aproximadamente).
En el eje vertical está la memoria utilizada en bytes.

### ¿Qué se observa?

La línea crece de forma lineal y constante conforme aumenta el tamaño del arreglo.

Esto significa que:

A mayor cantidad de elementos, mayor consumo de memoria.

El crecimiento es proporcional al tamaño del arreglo.

No hay picos bruscos ni aumentos exponenciales.


```python
plt.plot(df["Tamano"], df["Memoria_Bytes"])

plt.xlabel("Tamaño del arreglo")
plt.ylabel("Memoria (bytes)")
plt.title("Uso de Memoria")
plt.show()
```
